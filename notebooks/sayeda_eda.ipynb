```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
# This is a bivariate analysis evaluating the relationship between People in Need and Funding Received from 2010-2014
# specifically for Afghanistan
```

```python
#Load the data

humanitarian = pd.read_csv("../data/2010-2024 Humanitarian Needs and Funding - Raw Data.csv")
```

```python
humanitarian.head()
```

```python
#convert the column of values into integers

humanitarian['Value'] = humanitarian['Value'].str.replace(',', '').astype(int)
```

```python
# select only Afghanistan

afghanistan = humanitarian[humanitarian["Crisis Country"] == "Afghanistan"]
afghanistan
```

```python
# df interested in people in need

afghan_people_in_need = afghanistan[afghanistan['Metric'] == 'People in need']
afghan_people_in_need
```

```python
#grouped by year and sum of values per year
grouped_people_in_need = afghan_people_in_need.groupby('Year')['Value'].sum()

grouped_people_in_need
```

```python
# df for afghanistans funding requirements 
afghan_funding_requirements = afghanistan[afghanistan['Metric'] == 'Funding requirements']
afghan_funding_requirements
```

```python
#grouped by year and sum of values per year 
grouped_funding_requirements = afghan_funding_requirements.groupby('Year')['Value'].sum()

grouped_funding_requirements
```

```python
#merge data frames together on years

merged_afghanistan = pd.merge(grouped_people_in_need, grouped_funding_requirements, on='Year', how='left')
merged_afghanistan
```

```python
#rename the columns from Value_x to people in need and Value_y to funding requirements 

merged_afghanistan = merged_afghanistan.rename(columns={'Value_x': 'People in Need', 'Value_y': 'Funding Requirements'})
```

```python
sns.scatterplot(merged_afghanistan, x = "People in Need", y = "Funding Requirements", color = "royalblue")

plt.xlabel('People in Need')
plt.ylabel('Funding Requirements (USD)')
plt.title('People in Need vs Funding Requirements in Afghanistan 2010-2024')

plt.show()

pearsons_correlation = merged_afghanistan[['People in Need', 'Funding Requirements']].corr()
print(pearsons_correlation)
```

```python
#df for afghanistan funding recieved
afghan_funding_received = afghanistan[afghanistan['Metric'] == 'Funding received']
afghan_funding_received
```

```python
#grouped by year and sum of values per year 
grouped_funding_received = afghan_funding_received.groupby('Year')['Value'].sum()

grouped_funding_received
```

```python
# EDA on Syria (PIN vs Funding Requirements): first select syria

syria = humanitarian[humanitarian["Crisis Country"] == "Syria"]
syria
```

```python
# df interested in people in need

syrian_people_in_need =syria[syria['Metric'] == 'People in need']
syrian_people_in_need
```

```python
#group by year and summed values per year
grouped_syrian_people_in_need = syrian_people_in_need.groupby('Year')['Value'].sum()

grouped_syrian_people_in_need
```

```python
#df for funding requirements for syria 
syrian_funding_requirements =syria[syria['Metric'] == 'Funding requirements']
syrian_funding_requirements
```

```python
#group by year and summed values per year
grouped_syrian_funding_requirements = syrian_funding_requirements.groupby('Year')['Value'].sum()

grouped_syrian_funding_requirements
```

```python
#merge the grouped dfs
merged_syria = pd.merge(grouped_syrian_people_in_need, grouped_syrian_funding_requirements, on='Year', how='left')
merged_syria
```

```python
sns.scatterplot(merged_syria, x = "People in Need", y = "Funding Requirements", color = "royalblue")

plt.xlabel('People in Need')
plt.ylabel('Funding Requirements (USD)')
plt.title('People in Need vs Funding Requirements in Syria 2018-2024')

plt.show()

pearsons_correlation = merged_syria[['People in Need', 'Funding Requirements']].corr()
print(pearsons_correlation)
```

Question: What is the relationship between the amount of people in need and the funding requirements for Afghanistan from 2010 to 2024?

Answer: The bivariate analysis shows a *strong, positive* correlation between the amount of people in need and funding requirements for Afghanistan from 2010 to 2024. That is, as the amount of people in need increased over the years, so did the amount of funding requirements to support them. This is a very strong correlation, given that the pearson's correlation coefficient is 0.852583.
